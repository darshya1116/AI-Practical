```python
import matplotlib.pyplot as plt
import networkx as nx
import time

class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):
        self.graph.setdefault(u, []).append(v)
        self.graph.setdefault(v, []).append(u)

    def draw_graph(self, visited=None):
        G = nx.Graph()
        for u in self.graph:
            for v in self.graph[u]:
                G.add_edge(u, v)
        pos = nx.spring_layout(G)

        if visited is None:
            visited = set()

        node_color = ['lightgreen' if node in visited else 'lightblue' for node in G.nodes()]
        nx.draw(G, pos, with_labels=True, node_size=700, node_color=node_color, font_weight='bold')
        plt.title("DFS Traversal")
        plt.show()

    def dfs(self, start, visited=None):
        if visited is None:
            visited = set()

        visited.add(start)
        print(f"Visited: {start}")
        self.draw_graph(visited)
        time.sleep(1)

        for neighbor in self.graph.get(start, []):
            if neighbor not in visited:
                self.dfs(neighbor, visited)


def main():
    g = Graph()
    print("Enter 5 edges (format: u v):")
    for i in range(5):
        edge = input(f"Edge {i+1}: ")
        u, v = edge.split()
        g.add_edge(u, v)

    g.draw_graph()
    start_vertex = input("Enter starting vertex for DFS: ")
    print("\nDepth First Search:")
    g.dfs(start_vertex)


if __name__ == "__main__":
    main()
```
